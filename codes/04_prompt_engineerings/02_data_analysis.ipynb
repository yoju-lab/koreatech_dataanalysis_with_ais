## 프롬프트 in copilot
- total_bill과 tip 관계 분석 할 때 최적 방식 제시
- 두 그래프를 초보자가 알기 쉽게 설명
- 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
import numpy as np

# 데이터 로드
tips = sns.load_dataset("tips")

# 데이터 확인
print(tips.head())

# 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()

# 상관 분석
correlation, _ = pearsonr(tips['total_bill'], tips['tip'])
print(f'Pearson correlation: {correlation}')

# 회귀 분석
X = tips['total_bill'].values.reshape(-1, 1)
y = tips['tip'].values
reg = LinearRegression().fit(X, y)
predictions = reg.predict(X)

# 회귀선 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.plot(tips['total_bill'], predictions, color='red', linewidth=2)
plt.title('Total Bill vs Tip with Regression Line')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()

# 회귀 분석 결과
print(f'Intercept: {reg.intercept_}')
print(f'Slope: {reg.coef_[0]}')
print(f'R^2: {reg.score(X, y)}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
import numpy as np

# 데이터 로드
tips = sns.load_dataset("tips")

# 데이터 확인
print(tips.head())

# 상관 분석
correlation, _ = pearsonr(tips['total_bill'], tips['tip'])
print(f'Pearson correlation: {correlation}')

# 회귀 분석
X = tips['total_bill'].values.reshape(-1, 1)
y = tips['tip'].values
reg = LinearRegression().fit(X, y)
predictions = reg.predict(X)

# 1행 2열로 그래프 배치
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

# 첫 번째 그래프: 산점도
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axes[0])
axes[0].set_title('Total Bill vs Tip')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Tip')

# 두 번째 그래프: 회귀선이 포함된 산점도
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axes[1])
axes[1].plot(tips['total_bill'], predictions, color='red', linewidth=2)
axes[1].set_title('Total Bill vs Tip with Regression Line')
axes[1].set_xlabel('Total Bill')
axes[1].set_ylabel('Tip')

plt.tight_layout()
plt.show()
```

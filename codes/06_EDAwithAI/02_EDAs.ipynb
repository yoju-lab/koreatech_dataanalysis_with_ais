# Import Required Libraries
Import the necessary libraries, including seaborn and pandas.

```python
import seaborn as sns
import pandas as pd
```

# Load Titanic Dataset
Load the Titanic dataset using seaborn's load_dataset function.

```python
# Load Titanic Dataset
titanic_df = sns.load_dataset('titanic')

# Display the first few rows of the dataset
titanic_df.head()
```

# 데이터 개요 및 기본 정보 확인
Check the structure of the dataset using info(), describe(), head(), and tail() functions. Handle missing values and check data types for conversion.

```python
# Check the structure of the dataset
titanic_df.info()

# Display summary statistics of the dataset
titanic_df.describe()

# Display the first few rows of the dataset
titanic_df.head()

# Display the last few rows of the dataset
titanic_df.tail()

# Check for missing values
titanic_df.isnull().sum()

# Handle missing values (example: fill missing age with median)
titanic_df['age'].fillna(titanic_df['age'].median(), inplace=True)

# Check data types
titanic_df.dtypes

# Convert data types if necessary (example: convert 'survived' to category)
titanic_df['survived'] = titanic_df['survived'].astype('category')
```

# 변수별 분석
Analyze each variable including survived, pclass, sex, age, sibsp, parch, fare, embarked, embark_town, deck, class, who, adult_male, alone, and alive.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Analyze 'survived' variable
sns.countplot(x='survived', data=titanic_df)
plt.title('Survival Count')
plt.show()

# Analyze 'pclass' variable
sns.countplot(x='pclass', data=titanic_df)
plt.title('Passenger Class Count')
plt.show()

# Analyze 'sex' variable
sns.countplot(x='sex', data=titanic_df)
plt.title('Gender Count')
plt.show()

# Analyze 'age' variable
sns.histplot(titanic_df['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

# Analyze 'sibsp' variable
sns.countplot(x='sibsp', data=titanic_df)
plt.title('Number of Siblings/Spouses Aboard')
plt.show()

# Analyze 'parch' variable
sns.countplot(x='parch', data=titanic_df)
plt.title('Number of Parents/Children Aboard')
plt.show()

# Analyze 'fare' variable
sns.histplot(titanic_df['fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()

# Analyze 'embarked' variable
sns.countplot(x='embarked', data=titanic_df)
plt.title('Port of Embarkation')
plt.show()

# Analyze 'embark_town' variable
sns.countplot(x='embark_town', data=titanic_df)
plt.title('Town of Embarkation')
plt.show()

# Analyze 'deck' variable
sns.countplot(x='deck', data=titanic_df)
plt.title('Deck')
plt.show()

# Analyze 'class' variable
sns.countplot(x='class', data=titanic_df)
plt.title('Class')
plt.show()

# Analyze 'who' variable
sns.countplot(x='who', data=titanic_df)
plt.title('Who')
plt.show()

# Analyze 'adult_male' variable
sns.countplot(x='adult_male', data=titanic_df)
plt.title('Adult Male')
plt.show()

# Analyze 'alone' variable
sns.countplot(x='alone', data=titanic_df)
plt.title('Alone')
plt.show()
```

```python
# Analyze 'survived' variable
sns.barplot(x=titanic_df['survived'].value_counts().index, y=titanic_df['survived'].value_counts(), errorbar='sd')
plt.title('Survival Count')
plt.show()
```

# 변수 간 관계 분석
Perform correlation analysis, cross-analysis, and visualization to understand relationships between variables.

```python
# Perform correlation analysis
correlation_matrix = titanic_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Cross-analysis: Survival rate by passenger class
sns.catplot(x='pclass', hue='survived', data=titanic_df, kind='count', height=6, aspect=1.5)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Cross-analysis: Survival rate by gender
sns.catplot(x='sex', hue='survived', data=titanic_df, kind='count', height=6, aspect=1.5)
plt.title('Survival Rate by Gender')
plt.show()

# Cross-analysis: Survival rate by embarkation town
sns.catplot(x='embark_town', hue='survived', data=titanic_df, kind='count', height=6, aspect=1.5)
plt.title('Survival Rate by Embarkation Town')
plt.show()

# Cross-analysis: Survival rate by age group
titanic_df['age_group'] = pd.cut(titanic_df['age'], bins=[0, 12, 18, 35, 60, 80], labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
sns.catplot(x='age_group', hue='survived', data=titanic_df, kind='count', height=6, aspect=1.5)
plt.title('Survival Rate by Age Group')
plt.show()

# Visualize pairplot for selected variables
sns.pairplot(titanic_df[['age', 'fare', 'pclass', 'survived']], hue='survived', height=2.5)
plt.show()
```

# 인사이트 도출 및 가설 검증
Identify key factors, perform statistical tests, and select variables for modeling.

```python
from scipy import stats

# Identify key factors: Survival rate by passenger class and gender
sns.catplot(x='pclass', hue='sex', col='survived', data=titanic_df, kind='count', height=6, aspect=1.5)
plt.show()

# Perform statistical tests: Chi-square test for independence between 'pclass' and 'survived'
contingency_table = pd.crosstab(titanic_df['pclass'], titanic_df['survived'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-square test for 'pclass' and 'survived': chi2={chi2}, p-value={p}")

# Perform statistical tests: Chi-square test for independence between 'sex' and 'survived'
contingency_table = pd.crosstab(titanic_df['sex'], titanic_df['survived'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-square test for 'sex' and 'survived': chi2={chi2}, p-value={p}")

# Perform statistical tests: T-test for age difference between survived and not survived
survived_age = titanic_df[titanic_df['survived'] == 1]['age'].dropna()
not_survived_age = titanic_df[titanic_df['survived'] == 0]['age'].dropna()
t_stat, p_val = stats.ttest_ind(survived_age, not_survived_age)
print(f"T-test for age difference: t-statistic={t_stat}, p-value={p_val}")

# Select variables for modeling: Based on EDA and statistical tests
selected_features = ['pclass', 'sex', 'age', 'fare', 'embarked', 'sibsp', 'parch']

# Display selected features
print("Selected features for modeling:", selected_features)
```

# 추가적인 분석 방향
Explore feature engineering, time-related analysis, and text data analysis.

```python
# 추가적인 분석 방향

# 특성 공학: 가족의 크기(Family Size)라는 새로운 변수를 생성
titanic_df['family_size'] = titanic_df['sibsp'] + titanic_df['parch'] + 1
sns.countplot(x='family_size', data=titanic_df)
plt.title('Family Size Distribution')
plt.show()

# 가족 크기와 생존 여부의 관계 분석
sns.catplot(x='family_size', hue='survived', data=titanic_df, kind='count', height=6, aspect=1.5)
plt.title('Survival Rate by Family Size')
plt.show()

# 시간 관련 분석: 탑승 요일을 추출 (가정: 'embark_town'에 따라 임의로 요일 부여)
import numpy as np
np.random.seed(0)
titanic_df['embark_day'] = np.random.choice(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'], size=len(titanic_df))
sns.countplot(x='embark_day', data=titanic_df)
plt.title('Embarkation Day Distribution')
plt.show()

# 탑승 요일과 생존 여부의 관계 분석
sns.catplot(x='embark_day', hue='survived', data=titanic_df, kind='count', height=6, aspect=1.5)
plt.title('Survival Rate by Embarkation Day')
plt.show()

# 텍스트 데이터 분석: 이름 길이(Name Length)라는 새로운 변수를 생성
titanic_df['name_length'] = titanic_df['name'].apply(len)
sns.histplot(titanic_df['name_length'], bins=30, kde=True)
plt.title('Name Length Distribution')
plt.show()

# 이름 길이와 생존 여부의 관계 분석
sns.catplot(x='name_length', hue='survived', data=titanic_df, kind='count', height=6, aspect=1.5)
plt.title('Survival Rate by Name Length')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터셋 로드
titanic_df = sns.load_dataset('titanic')

# 성별 생존자 수
sns.countplot(x='sex', hue='survived', data=titanic_df)
plt.title('Survival Count by Gender')
plt.show()

# 객실 등급별 생존자 수
sns.countplot(x='class', hue='survived', data=titanic_df)
plt.title('Survival Count by Class')
plt.show()
```

```python
import matplotlib.pyplot as plt
# 연령 분포
plt.hist(titanic_df['age'].dropna(), bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 운임 분포
sns.kdeplot(titanic_df['fare'].dropna())
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
```

```python
# 형제/자매/배우자 수 분포
sns.countplot(titanic_df['sibsp'])
plt.title('Sibling/Spouse Count Distribution')
plt.xlabel('Number of Siblings/Spouses')
plt.ylabel('Frequency')
plt.show()

# 부모/자녀 수 분포
sns.countplot(titanic_df['parch'])
plt.title('Parent/Children Count Distribution')
plt.xlabel('Number of Parents/Children')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
# 연령 구간 생성
titanic_df['age_group'] = pd.cut(titanic_df['age'], bins=range(0, 90, 10))

# 연령 구간별 생존자 수
sns.countplot(x='age_group', hue='survived', data=titanic_df)
plt.title('Survival Count by Age Group')
plt.xlabel('Age Group (10-year intervals)')
plt.ylabel('Count')
plt.show()
```

## 탑승 위치에 따른 생존율:

탑승 위치별 생존율: 승객이 탑승한 위치(Embarked)에 따른 생존율을 분석합니다.

```python
# 탑승 위치별 생존율
sns.barplot(x='embarked', y='survived', data=titanic_df)
plt.title('Survival Rate by Embarked Location')
plt.show()
```
